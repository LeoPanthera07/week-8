```python
import pandas as pd
import numpy as np

df = pd.read_csv('train_data.csv')
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()
df.duplicated().sum()
```

```python
df['target'] = df['Stay'].apply(lambda x: 1 if x in ['0-10','11-20','21-30'] else 0)
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col].fillna(df[col].median(), inplace=True)

for col in cat_cols:
    df[col] = df[col].astype(str).str.lower().str.strip()
    df[col].fillna(df[col].mode()[0], inplace=True)

df = df.drop_duplicates()
```

```python
df['age_group'] = df['Age'].apply(lambda x: 0 if float(x[-2:]) < 30 else 1 if float(x[-2:]) < 60 else 2)

df['severity_flag'] = df['Severity of Illness'].map({
    'minor': 0,
    'moderate': 1,
    'extreme': 2
})

df['resource_usage'] = df['Visitors with Patient'] * df['Admission_Deposit']
```

```python
df = df.drop(['Stay'], axis=1)
```

```python
df.to_csv('clean_data.csv', index=False)
```

- Target derived from Stay to reflect 30-day readmission risk
- Missing values handled using median/mode for robustness
- Age grouped to reduce noise and capture nonlinear trends
- Severity encoded ordinally to preserve ranking
- Interaction feature captures hospital resource usage

# **Medium**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('clean_data.csv')
```

```python
TARGET = 'target'
ID_COLS = ['case_id','patientid']

X = df.drop(ID_COLS + [TARGET], axis=1, errors='ignore')
y = df[TARGET]
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=300, class_weight='balanced')
lr.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=150, class_weight='balanced')
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import f1_score, roc_auc_score

lr_preds = lr.predict(X_test)
rf_preds = rf.predict(X_test)

print("LR F1:", f1_score(y_test, lr_preds))
print("RF F1:", f1_score(y_test, rf_preds))

print("LR AUC:", roc_auc_score(y_test, lr.predict_proba(X_test)[:,1]))
print("RF AUC:", roc_auc_score(y_test, rf.predict_proba(X_test)[:,1]))
```

- Stratified split ensures class distribution consistency
- F1 chosen due to imbalance and clinical importance of recall
- AUC evaluates ranking ability of model
- Logistic Regression used as baseline
- Random Forest captures nonlinear relationships
- Final model selected based on combined F1 and AUC performance
